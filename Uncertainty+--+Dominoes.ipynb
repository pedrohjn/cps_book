# Drawning one dominoes tile from the deck

## Case 1: Chance to get each one of the 28 tiles

```python
snapshots = 10000
drawing2 = np.random.randint(1,29, size=snapshots)
```

```python
plt.figure(figsize=(16,7))
plt.hist(drawing2, bins=28, rwidth=0.7)
plt.plot([1,28], [snapshots/28,snapshots/28])
#plt.ylim([0, 1])
plt.xlim([1, 28])
plt.title('Frequency diagram for one-tile drawing in a dominoes set', size=18)
plt.xlabel('\n Tiles', size=18)
plt.ylabel('Number of observations out of ' +str(snapshots),size=18)
plt.xticks([])
plt.annotate('[0,0]', xy=(0,0), xytext=(1,-20), size=14,annotation_clip=False)
plt.annotate('[0,1]', xy=(0,0), xytext=(2,-20), size=14,annotation_clip=False)
plt.annotate('...', xy=(0,0), xytext=(4,-20), size=14,annotation_clip=False)
plt.annotate('...', xy=(0,0), xytext=(25.5,-20), size=14,annotation_clip=False)
plt.annotate('[6,6]', xy=(0,0), xytext=(27,-20), size=14,annotation_clip=False)
plt.savefig('hist_1.pdf',pad_inches=0) 
plt.show()
```

```python
plt.figure(figsize=(16,7))
plt.hist(drawing2, bins=28, rwidth=0.7, density = True)
plt.plot([1,28], [1/28,1/28])
#plt.ylim([0, 1])
plt.xlim([1, 28])
plt.title('Probability mass function: Empirical and analytic values', size=18)
plt.xlabel('\n$x$', size=18)
plt.ylabel('$P\;(x)$',size=18)
plt.xticks([])
plt.annotate('1', xy=(0,0), xytext=(1.25,-20/10000), size=14,annotation_clip=False)
plt.annotate('2', xy=(0,0), xytext=(1.25 + 1.1,-20/10000), size=14,annotation_clip=False)
plt.annotate('...', xy=(0,0), xytext=(1.25 + 1.1*2,-20/10000), size=14,annotation_clip=False)
plt.annotate('...', xy=(0,0), xytext=(1.25 + 1.1*22,-20/10000), size=14,annotation_clip=False)
plt.annotate('28', xy=(0,0), xytext=(1.25 + 26,-20/10000), size=14,annotation_clip=False)
plt.title('Probability mass function: Empirical and analytic values', size=18)
plt.savefig('pmf_1.pdf',pad_inches=0) 
plt.show()
```

## Case 3: Sum of numbers

```python
snapshots = 10000
drawing3 = np.random.randint(7, size=(snapshots, 2))
sum_value=np.zeros(snapshots)
for i in range(snapshots):
    sum_value[i]= drawing3[i,0] + drawing3[i,1]
    i=i+1
```

## Arrival as Poisson

```python
##
n = np.arange(140)
plt.figure(figsize=(16,7))
plt.step(n, poisson.pmf(n, 5), where='mid', label='5 arrivals per hour')
plt.step(n, poisson.pmf(n, 20), where='mid', label='20 arrivals per hour')
plt.xlabel('Number of arrivals $x$', size=18)
plt.ylabel(' Probability of $x$ arrivals in \n the period, i.e., $P\;(X = x)$', size=18)
plt.title('Poisson distribution', size=18)
#plt.ylim([0, 1])
plt.xlim([0, 35])
plt.grid(True) #grid
plt.legend(fontsize=14)
plt.savefig('PMF_Poisson.pdf',pad_inches=0) 
plt.show()
```

## Maxwell-Boltzmann [probability density function](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.maxwell.html#scipy.stats.maxwell)

Micro and macro states

```python
x = np.linspace(maxwell.ppf(0), maxwell.ppf(0.99), 100)
thres=x[np.argmax(maxwell.pdf(x))]
plt.figure(figsize=(14,9))
plt.plot(x,maxwell.pdf(x))
plt.axvline(thres,linestyle=':',color='k')
#plt.plot(n, maxwell.pdf(n, 20), label='20 arrivals per hour')
plt.xlabel('$x$', size=18)
plt.ylabel('$f_X(x)$', size=18)
plt.title('Maxwell-Boltzmann distribution', size=18)
plt.ylim([0, 1])
plt.xlim([0, np.max(x)])
plt.annotate('Demons threshold', xy=(0,0), xytext=(thres + 0.05,0.9), size=16,annotation_clip=False)
plt.grid(True) #grid
#plt.xticks([])
#plt.yticks([])
#plt.legend(fontsize=14)
plt.savefig('PDF_Maxwell.pdf',pad_inches=0) 
plt.show()
```

## Queue >> stochastic process

```python
#time periods: 10:00 - 14:30
size=10
#Arrival
mean=1.05
arrival = np.random.poisson(mean, size)
#Service
mean=1
service = np.random.poisson(mean, size)
#
queue = np.zeros(size+1)
n = np.arange(size+1)
for i in range(size):
    if ((arrival[i] - service[i]) + queue[i]) > 0:
        queue[i+1] = max(0,(arrival[i] - service[i]) + queue[i])

###Day 2
#Arrival
mean=1.05
arrival = np.random.poisson(mean, size)
#Service
mean=1
service = np.random.poisson(mean, size)
queue2 = np.zeros(size+1)
for i in range(size):
    if ((arrival[i] - service[i]) + queue2[i]) > 0:
        queue2[i+1] = max(0,(arrival[i] - service[i]) + queue2[i])
###Day 3
#Arrival
mean=1.05
arrival = np.random.poisson(mean, size)
#Service
mean=1
service = np.random.poisson(mean, size)
queue3 = np.zeros(size+1)
for i in range(size):
    if ((arrival[i] - service[i]) + queue3[i]) > 0:
        queue3[i+1] = max(0,(arrival[i] - service[i]) + queue3[i])

plt.figure(figsize=(16,4))
plt.step(n, queue, linestyle='dashdot',color='k', label='Day 1')
plt.step(n, queue2,  linestyle='--', label='Day 2')
plt.step(n, queue3,  linestyle='-', label='Day 3')
for i in range(size + 1):
    plt.axvline(i,linestyle=':',color='wheat');

plt.xlabel('Time index $k$', size=18)
plt.ylabel('Person in queue: $X(k)$', size=18)
plt.title('Persons lined up in a restaurant queue', size=18)
plt.ylim([0, 10])
plt.xlim([-0.2, 10])
plt.xticks(np.arange(10),['10:00','10:30','11:00','11:30','12:00','12:30','13:00','13:30','14:00', '14:30'])
plt.grid(True) #grid
plt.legend(fontsize=14)
plt.savefig('queue.pdf',pad_inches=0) 
plt.show() 
```
